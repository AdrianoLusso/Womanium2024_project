### 2.1 Model creation

First, we create the $\textbf{function predictor}$, which make use of a quantum circuit to modelate the behaviour of a numerical function $f(x)$. It is formed by the sum between a bias and the expectation value of Pauli Z with respect a variational circuit, parametrized by some weights and x-values.

An $f(x)$ is formed by a set of x-values, domain values, or domain, that are mapped to a set of y-values, or range values, or range. Knowing this, the 1-qubit variational circuit is formed by:
- an initialization using the $H$ gate, so as to use the Hadamard basis. This will help, in the following layers, to cuantify the x-values using the phase of the qubit.
- $n$ layers, where each one is formed by:
    - an $RZ(x)$ gate, for cuantifying the x-value using the phase of the qubit.
    - an $RX(x)$ gate, for cuantifying the y-value using the amplitude of the qubit.

For a sine numerical function, just a 1-layer variational circuit is enough, but while the numerical function is getting more complex, a bigger number of layers will be necessary for adapting to its custom behaviour.

```python
def function_predictor(weights, bias, x):
    return circuit(weights, x) + bias

dev = qml.device("default.qubit")
@qml.qnode(dev)
def circuit(weights, x):
    qml.Hadamard(wires=0)

    for layer_weights in weights:
        layer(layer_weights,x)
    
    return qml.expval(qml.PauliZ(0))

def layer(weights,x):
    qml.RZ(x + weights[0],wires=0)
    #qml.RX(weights[1] * np.pi,wires=0)
    qml.RX(weights[1],wires=0)
```

Then, we define a cost function. This is useful for characterizing how well a set of parameters works with the function predictor. The cost function will do the prediction for a particular set of x-values, weights and bias. After that , it calculates the cost comparing the predictions with the current y-values associated to the x-values.

The cost comparison is done with the Root Mean Squared Error (RMSE) metric, which formula is

$\sqrt{ \tfrac{\sum_{i=1}^{N}(y_i - \hat{y}_i)^2  }{N} },$

where $N$ is the amount of y-values/predictions, $y_i$ is the $i_{th}$ y-value and $\hat{y}_i$ is its the corresponding prediction. RMSE is commonly used for standarized data, and since the function to evaluate will be standarize to real values $Y \in [-1,1]$, it is a well choise. Also, compared against its simpler version Mean Squared Error (MSE), or other metrics like Mean Absolute Percentual Error (MAPE), it deal better with the outliers, or with y-values too close to 0.

```python
def cost(weights, bias, X, Y):
    predictions = [function_predictor(weights, bias, x) for x in X]
    return rmse(Y, predictions)

def rmse(labels, predictions):
    return np.sqrt(np.mean((labels - qml.math.stack(predictions)) ** 2))
```

Finally, we define a custom wrapper for an optimizer. This component make use of the function predictor and the cost function to decide how to variate the parameters, in order to minimize the difference between the predictions and the actual y-values. During the workflow, two different optimizers wil be used, and will be further explained in the next sections. The following cell just create a wrapper for one of those optimizers: $\textit{RotosolveOptimizer}$. The wrapper encapsulates the whole configuration of this optimizer, so as to facilitate the way of using it during the optimization of the variational parameters.

```python
class CustomRotosolveOptimizer():
    def __init__(self,nums_frequency, substep_optimizer='brute' , num_steps=4):
        self.nums_frequency = nums_frequency
        self.optimizer = RotosolveOptimizer(
            substep_optimizer = substep_optimizer,
            substep_kwargs={'num_steps':num_steps}
        )
        
    def step(self,cost,weights,bias,X,Y):
        weights , bias = self.optimizer.step(cost, weights, bias, X=X, Y=Y, nums_frequency= self.nums_frequency)   
        return weights , bias
```

### 2.2 Data preparation

The following numerical functions will be the ones the function predictor should learn:

1. A regular sine function.
2. A sine function which is not naturally limited to $Y \in [-1,1]$.
3. A regular cosine function.
4. A sum of different sine and cosine functions.
5. A sum of different and more complex sine and cosine functions.

```python
def func1(X):
    return np.sin(X)

def func2(X):
    return 23 * np.sin(X) -7

def func3(X):
    return np.cos(X)

def func4(X):
    return np.sin(X) + np.cos(X) + np.sin(3*X)

def func5(X):
    return np.sin(X) + 2*np.cos(4*X) + 0.5*np.sin(-3*X) + np.cos(10*X)
funcs = [func1,func2,func3,func4,func5]
```

As mentioned before, some of the numerical functions need to be normalize to $Y \in [-1,1]$ for the training, and after that, de-normalize for evaluating the final accuracy of the fitting. For that, the following functions had been created:

```python
def normalize(Y,min_value,max_value):
    return  -1 + 2 * (Y - min_value) / (max_value - min_value)
    
def denormalize(Y,min_value,max_value):
    return min_value + (Y + 1) * (max_value - min_value) / 2
```

Now, lets start with the data prepration. The first step involves creating the numerical functions, so we will:
1. Define the continous X-axis from 0 to $2\pi$.
2. Define the size of the dataset. That is, the amount of points that discretize the numerical function.
    - The amount of points chosen depends of the complexity of the numerical function.
3. For each of the 5 numerical functions:
    - Calculate the  the y-values for each x-value point.
    - calculate the minimum and maximum value of the not-normalize Y-axis.
    - Using the min. and max. values, normalize the numerical function range to $Y \in [-1,1]$.
        - As said before,the normalized numerical functions will be the one used for the training.
        - It is important to save the min. and max. values for future de-normalization needed. 

```python
# limit the x-axis from 0 to 2 * np.pi
a = 0
b = 2 * np.pi

# generate 'n_points'amount of equidistant points in the range [2,2*pi]
n_points = [100,100,100,100,500] 
Xs = [np.linspace(a, b, n_p) for n_p in n_points]

Ys = []
minmax_values = []
fig, axs = plt.subplots(1, 5, figsize=(25, 3)) 
for itr,func in enumerate(funcs):
    #calculate the numerical function
    X = Xs[itr]
    Y = func(X)

    # create the plot for the numerical function
    axs[itr].scatter(X, Y,s=10, label='y')
    axs[itr].set_title('not-normalized numerical function '+str(itr+1))
    axs[itr].set_xlabel('x')
    axs[itr].set_ylabel('y')
    axs[itr].grid(True)

    # write the x-axis in terms of radians
    radian_ticks = np.linspace(0, 2 * np.pi, 5)  
    radian_labels = ['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']
    axs[itr].set_xticks(radian_ticks, radian_labels)

    # calculate the min and max values of the numerical func, so as to normalize it
    min_value = np.min(Y)
    max_value = np.max(Y)
    Y = normalize(Y,min_value,max_value)

    # save the normalized numerical func and its min and max values
    minmax_values.append((min_value,max_value))
    Ys.append(Y)
```

The second step goes about defining the training and test set, where the training set is the $75\%$ of the total dataset. The division will be plot, so that the blue points are for training and the orange points are for test.


```python
np.random.seed(0)

# for each function, define the size of the training set
nums_train = [int(0.75 * n_p) for n_p in n_points]

# define indexes for the train and test set for sample size of 100 (for funcs 1,2,3,4) and 500 (for func 5)
index_100 = np.random.permutation(range(n_points[0]))
indexes = [
index_100,
index_100,
index_100,
index_100,
np.random.permutation(range(n_points[4]))
]

# for each function, create the training and test sets (Ys still empty)
Ys_train = []
Ys_test = []

Xs_train = [ Xs[itr][indexes[itr][:num_train]] for itr,num_train in enumerate(nums_train) ]
Xs_test = [ Xs[itr][indexes[itr][num_train:]] for itr,num_train in enumerate(nums_train) ]

fig, axs = plt.subplots(1, 5, figsize=(25, 3)) 
for itr,Y in enumerate(Ys):
    # auxiliary data for a function
    #n_p = n_points[itr]
    num_train = nums_train[itr]
    X_train = Xs_train[itr]
    X_test = Xs_test[itr]
    index = indexes[itr]
    
    # create the test sets    
    Y_train = Y[index[:num_train]]
    Y_test = Y[index[num_train:]]
    
    # create the plot for the numerical function
    axs[itr].scatter(X_train, Y_train,s=10, label='y')
    axs[itr].scatter(X_test, Y_test,s=10, label='y')
    axs[itr].set_title('normalized numerical function '+str(itr+1))
    axs[itr].set_xlabel('x')
    axs[itr].set_ylabel('y')
    axs[itr].grid(True)

    # write the x-axis in terms of radians
    radian_ticks = np.linspace(0, 2 * np.pi, 5)  # Cambia el 5 por el nÃºmero de etiquetas que quieras
    radian_labels = ['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']
    axs[itr].set_xticks(radian_ticks, radian_labels)

    #save the Y train and test sets
    Ys_train.append(Y_train)
    Ys_test.append(Y_test)
```

### 2.3 Workflows creation

In this section, there will be created three functions that will be of help during the training and evaluation steps.

First, a function that encapsulated the whole optimization of the variational parameters for a particular numerical function. It receives as parameters:
* The initial parameters structure
* The training and test sets
* The optimizer to be used
* The number of iterations/epochs of the optimization

For each step of the optimization, the next will happen:
1. the optimizer will be run one step, giving us a a change over the weights and bias
2. using this new parameters, it will be done the predictions over the X train set and the X test set.
3. The cost for the X train set predictions and the corresponding labels will be calculated.
    * if this is the new minimal cost, all the information of this iteration is saved for beeing return at the end of the optimization
4. The RMSE between the X test set predictions and its labels is calculated.

```python
#def optimize(inits,X_train,Y_train,X_test,Y_test,nums_frequency,opt_itrs=100):
def optimize(inits,X_train,Y_train,X_test,Y_test,optimizer,opt_itrs=100):
    #opt_kwargs = {"num_steps": 4}
    #optimizer = RotosolveOptimizer(substep_optimizer="brute", substep_kwargs=opt_kwargs)
    #optimizer = NesterovMomentumOptimizer(0.02)
    
    weights = inits[0]
    bias = inits[1]

    best_weights = None
    best_bias = None
    best_itr = None
    best_cost = float('inf')
    costs = []
    
    for itr in range(opt_itrs):
        #weights, bias = optimizer.step(cost, weights, bias, X=X_train, Y=Y_train,nums_frequency=nums_frequency)
        weights, bias = optimizer.step(cost, weights, bias, X=X_train, Y=Y_train)
        
        # Compute the predictions
        #predictions = [np.sign(function_predictor(weights, bias, x)) for x in X]
        predictions_train = [function_predictor(weights,bias,x) for x in X_train]
        predictions_test = [function_predictor(weights,bias,x) for x in X_test]
    
        #compute the current cost
        current_cost = cost(weights, bias, X_train, Y_train)
        if current_cost < best_cost:
            best_itr = itr+1
            best_cost = current_cost
            best_weights = weights
            best_bias = bias
        costs.append(current_cost)
        
        # compute the accuracy for the training set. This will show how well is
        # the classifier classifyng the training set
        acc_test = rmse(Y_test, predictions_test)

        if itr+1 == 1 or (itr + 1) % 5 == 0:
            #print(weights)
            #print(bias)
            print(f"Iter: {itr+1:4d} | Cost (train_RMSE): {current_cost:0.7f} | test_RMSE: {acc_test:0.7f}")
    return best_weights,best_bias,best_itr,best_cost,costs
```

Then, we define the evaluation function, where given a set of parameters, it returns an evaluation analysis for them. For each numerical function:

* plot the RMSE. For this, the function keeps normalized, so as to have a relativized way of comparing this metric.
* plots its y-values and te corresponding predictions. For this, the function is denormalized. This gives a more visual way of analyzing the accuracy of the function predictor.

```python
def evaluate(params):
    # for each numerical function, calculate the predictions over the parameters
    predictions = []
    for itr, param in enumerate(params):
        function_predictions = []
        X = Xs_test[itr]
        for x in X:
            prediction=function_predictor(param[0],param[1],x)
            function_predictions.append(prediction)


    
    #for itr,X in enumerate(Xs_test):
    #    function_predictions = []
    #    for x in X:
    #        prediction=function_predictor(params[itr][0],params[itr][1],x)
     #       function_predictions.append(prediction)

        # before saving the predictions, they are denormalized using the min. and max. values corresponding to each original numerical function
        min,max = minmax_values[itr]
        predictions.append(   denormalize(qml.math.stack(function_predictions),min,max)   )

    fig, axs = plt.subplots(1, len(predictions), figsize=(25, 3)) 
    for itr,f_predictions in enumerate(predictions):
    #for itr,Y in enumerate(Ys_test):
        # auxiliary data for a function
        n_p = n_points[itr]
        num_train = nums_train[itr]
        X_test = Xs_test[itr]
        index = indexes[itr]
        #f_predictions = predictions[itr]
        
        # create the denormalized test sets    
        min,max = minmax_values[itr]
        Y_test = denormalize(Ys_test[itr],min,max)

        # create the plot for the numerical function
        axs[itr].scatter(X_test, Y_test,s=10,alpha=0.5, label='labels')
        axs[itr].scatter(X_test, f_predictions,s=10, alpha=0.5,label='predictions')
        axs[itr].set_title('func'+str(itr+1))
        axs[itr].set_xlabel('x')
        axs[itr].set_ylabel('y')
        axs[itr].grid(True)
    
        # write the x-axis in terms of radians
        radian_ticks = np.linspace(0, 2 * np.pi, 5)  # Cambia el 5 por el nÃºmero de etiquetas que quieras
        radian_labels = ['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']
        axs[itr].set_xticks(radian_ticks, radian_labels)

        axs[itr].legend()

        #  plot the RMSE, showing how much error it has the prediction
        print('RMSE Numerical function',itr,': ',rmse(Y_test,f_predictions))
```

Now, its time to explain the optimizer to be used and the parameters needed for its creation.

As mentioned before, the optimizers to be used are $\textit{RotosolveOptimizer}$ [3] and $\textit{NesterovMomentumOptimizer}$ [4]. The former is gradient-free optimizer, which altough converges slower than gradient-based optimizers, is better suited to NISQ devices [2]. It works by associating each parameter to a corresponding rotation gate. The optimizer will update each parameter independently, while keeping the other fixed. This optimizer is recommended for easy-to-evaluate functions where there are few parameters to optimize. Due to this, $\textit{RotosolveOptimizer}$ will be used for optimizing the function predictor over numerical functions 1, 2, 3 and 4. Since the last numerical function has a larger set of parameters, it won't be used with this optimizer.

The $\textit{RotosolveOptimizer}$ makes use of three parameters for it to work:
* "substep_optimizer", which defines an optimizer to use for the single-parameter global optimization . By default set with brute force.
* "num_steps", by default set as "4".
* "nums_frequency", which defines the periodicity in which a particular parameter affects the cost. All paramaters has a periodicity of 1. The dictionary containing this information is created in the following code cell.

The second one is a gradient-based optimizer, which update the parameters moving along the direction indicated by the gradient at the specific points. The most known gradient-based optimizer is the $\textit{GradientDescentOptimizer}$ [5]. In particular, $\textit{NesterovMomentumOptimizer}$ makes use of a momentum term for defining the value of a parameter $x$ at the next iteration $t+1$:

$x^{(t+1)} = x^{(t)} - a^{(t+1)}$.

The accumulator term $a$  defined as:

$a^{(t+1)} = ma^{(t)} + \eta \nabla f (x^{(t)} - ma^{(t)}),$

where $m$ is the momentum and $\eta$ the step size.

For situations where the amount of variables increase, this kind of optimizer is recommended. Due to this, it will be used for optimizing the function predictor not just on the first four numerical functions, but also for the fifth one.

The $\textit{NesterovMomentumOptimizer}$ parameter that need for it to work is the "size_step". The bigger the value of the parameter, the bigger the size of the step taken at each iteration. A good step size is one that is bigger enough to converge as fast as possible, but smaller enough to not get stuck at local minimals. For a bigger set of parameters to optimize, the smaller the step should be.

```python
sizes_steps = [0.2 , 0.2 , 0.2 , 0.05 , 0.05]

nums_frequences = []
for itr,init in enumerate(inits[0:4]):
    nums_frequency = {'weights':{},'bias':{():1}}
    weights = init[0]
    indexes = list(np.ndindex(weights.shape))
    for index in indexes:
        nums_frequency['weights'][index] = 1
    nums_frequences.append(nums_frequency)
    print('frequences for numerical function ',itr+1,': ')
    display(nums_frequency)
    print('\n')
```
